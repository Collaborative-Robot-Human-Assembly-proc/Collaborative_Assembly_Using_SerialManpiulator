```python
import sys

sys.path.append("/usr/lib/freecad/lib")
sys.path.append("/usr/lib/freecad/Mod")

import pyvista as pv
import FreeCAD as App
import FreeCADGui as Gui
import Part
import Mesh

import pyvista as pv
import xlwt

def m(step_file):
    shape = Part.Shape()
    shape.read(step_file)
    doc = App.newDocument('Doc')
    pf = doc.addObject("Part::Feature","MyShape")
    pf.Shape = shape
    Mesh.export([pf], 'manga.stl')

def extract_part_parameters(stp_file_path, output_file_path):
    # Start FreeCAD in console mode
    App.ConsoleMode = True
    App.newDocument()
    shape = Part.Shape()
    shape.read(step_file)
    doc = App.newDocument('Doc')
    # Import the STEP file
    
    # Retrieve the imported part object
    # part_object = doc.Objects[0]

    # Create an Excel workbook and worksheet
    dimensions = shape.BoundBox.DiagonalLength, shape.BoundBox.XLength, shape.BoundBox.YLength, shape.BoundBox.ZLength
    print(shape)

    # # Print the dimensions
    # print("Diagonal Length:", dimensions[0])
    # print("X Length:", dimensions[1])
    # print("Y Length:", dimensions[2])
    # print("Z Length:", dimensions[3])

    workbook = xlwt.Workbook()
    worksheet = workbook.add_sheet("Part_Parameters")

    # Iterate over the parameters of the part object and write them to the worksheet
    row = 0
    for param in doc.PropertiesList:
        worksheet.write(row, 0, param)
        worksheet.write(row, 1, str(doc.getPropertyByName(param)))
        row += 1

    # Save the Excel file
    workbook.save(output_file_path)
    print("Part parameters extracted and saved successfully.")

    # Close the document
    App.closeDocument(doc.Name)

    # Stop FreeCAD
    App.ConsoleMode = False

def import_step_and_visualize(stl_file):
    # Import the STEP file using pyservoce
    shape = pv.read(stl_file)

    # Convert the shape to a pyvista PolyData object
    shape.plot()

if __name__ == '__main__':
	# Usage example
    step_file = 'banhrang.stp'
    output_file_path = "parameters.xls"
    stl_file = "manga.stl"
    m(step_file)
    extract_part_parameters(step_file, output_file_path)
    import_step_and_visualize(stl_file)
```

```python
import pandas as pd

class A():
  def __init__(self):
     self.x1 = 'Attributes'
     self.x2 = 2
     self.x3 = 3
     self.x4 = 4
     self.y1 = 'Values'
     self.y2 = 2
     self.y3 = 3
     self.y4 = 4

  def access_attributes(self):
     data = []
     attributes = dir(self)  # Get a list of all attributes of the class object
     for attr_name in attributes:
         if attr_name.startswith('x') or attr_name.startswith('y'):
             try:
                 attr_value = getattr(self, attr_name)  # Get the attribute value
                 data.append([attr_name, attr_value])
             except AttributeError:
                 continue

     attribute_names = [x for x in attributes if x.startswith('x') or x.startswith('y')]
     df = pd.DataFrame(data, columns=attribute_names)
     print(df)

# Create an instance of class A
a = A()

# Call the method to access the attributes and store in a DataFrame
a.access_attributes()
```
