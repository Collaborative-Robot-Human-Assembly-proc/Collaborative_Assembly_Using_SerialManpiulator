```python
import tkinter as tk
import customtkinter as ctk
import subprocess
import sys

class App(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Terminal Output")
        self.geometry("400x300")

        self.main_frame = ctk.CTkFrame(self)
        self.main_frame.pack(side=ctk.TOP, fill=ctk.BOTH, expand=True)

        self.button_frame = ctk.CTkFrame(self.main_frame)
        self.button_frame.pack(side=ctk.TOP, pady=10)

        self.terminal_frame = ctk.CTkFrame(self.main_frame)
        self.terminal_frame.pack(side=ctk.TOP, fill=ctk.BOTH, expand=True)

        self.label_frame = ctk.CTkFrame(self.main_frame)
        self.label_frame.pack(side=ctk.TOP, pady=10)

        self.execute_button = ctk.CTkButton(self.button_frame, text="Execute", command=self.execute_command)
        self.execute_button.pack()

        self.terminal_output = TerminalOutput(self.terminal_frame)
        self.terminal_output.pack(side=ctk.LEFT, fill=ctk.BOTH, expand=True)

        self.scrollbar = ctk.CTkScrollbar(self.terminal_frame, command=self.terminal_output.yview)
        self.scrollbar.pack(side=ctk.RIGHT, fill=ctk.Y)

        self.terminal_output.config(yscrollcommand=self.scrollbar.set)

        self.manga_label = ctk.CTkLabel(self.label_frame, text="MANGA")
        self.manga_label.pack()

        sys.stdout = self.terminal_output
        sys.stderr = self.terminal_output

    def execute_command(self):
        command = 'ls'
        output = subprocess.check_output(command, shell=True, universal_newlines=True)
        self.terminal_output.write(output)

class TerminalOutput(tk.Text):
    def __init__(self, master, **kwargs):
        super().__init__(master, **kwargs)
        self.configure(state='disabled')

    def write(self, text):
        self.configure(state='normal')
        self.insert(ctk.END, text)
        self.configure(state='disabled')

    def flush(self):
        pass

app = App()
app.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog
import cv2
from PIL import Image, ImageTk

class VideoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("OpenCV Video")
        self.root.geometry("800x400")  # Set the window size
        
        # Create a label to display the video feed
        self.video_label = tk.Label(self.root)
        self.video_label.pack()
        
        # Initialize the video capture
        self.video_capture = None
        
        # Prompt user to select a video file
        self.select_video_file()

        # Start the video feed display
        self.update_video_feed()

    def select_video_file(self):
        # Prompt the user to select a video file
        file_path = filedialog.askopenfilename(filetypes=[("Video files", "*.mp4")])
        
        # Open the selected video file
        self.video_capture = cv2.VideoCapture(file_path)

    def update_video_feed(self):
        # Read a frame from the video capture
        ret, frame = self.video_capture.read()

        if ret:
            # Resize the frame to 800x400 pixels
            resized_frame = cv2.resize(frame, (800, 400))
            
            # Convert the frame from BGR to RGB
            rgb_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2RGB)
            
            # Create an ImageTk object from the RGB frame
            img = Image.fromarray(rgb_frame)
            img_tk = ImageTk.PhotoImage(image=img)
            
            # Update the video label with the new frame
            self.video_label.configure(image=img_tk)
            self.video_label.image = img_tk  # Keep a reference to prevent garbage collection
            
        # Schedule the next update after 10 milliseconds
        self.root.after(10, self.update_video_feed)

# Create the Tkinter root window
root = tk.Tk()

# Create an instance of the VideoApp
video_app = VideoApp(root)

# Start the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

style = ttk.Style()
style.theme_use("default")
style.configure("Treeview",
                background="#2a2d2e",
                foreground="white",
                rowheight=25,
                fieldbackground="#343638",
                bordercolor="#343638",
                borderwidth=0)
style.map('Treeview', background=[('selected', '#22559b')])
style.configure("Treeview.Heading",
                background="#565b5e",
                foreground="white",
                relief="flat")
style.map("Treeview.Heading",background=[('active', '#3484F0')])

root = tk.Tk()
root.title('Treeview demo')
root.geometry('620x200')

# define columns
columns = ('first_name', 'last_name', 'email')

tree = ttk.Treeview(root, columns=columns, show='headings')

# define headings
tree.heading('first_name', text='First Name')
tree.heading('last_name', text='Last Name')
tree.heading('email', text='Email')

# generate sample data
contacts = []
for n in range(1, 100):
    contacts.append((f'first {n}', f'last {n}', f'email{n}@example.com'))

# add data to the treeview
for contact in contacts:
    tree.insert('', tk.END, values=contact)


def item_selected(event):
    for selected_item in tree.selection():
        item = tree.item(selected_item)
        record = item['values']
        # show a message
        showinfo(title='Information', message=','.join(record))


tree.bind('<<TreeviewSelect>>', item_selected)

tree.grid(row=0, column=0, sticky='nsew')

# add a scrollbar
scrollbar = ttk.Scrollbar(root, orient=tk.VERTICAL, command=tree.yview)
tree.configure(yscroll=scrollbar.set)
scrollbar.grid(row=0, column=1, sticky='ns')

# run the app
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
import pandas as pd

def show_dataframe():
    # Create a sample DataFrame
    data = {'Name': ['John', 'Emma', 'Peter'],
            'Age': [25, 30, 35],
            'City': ['New York', 'London', 'Paris']}
    df = pd.DataFrame(data)
    
    # Create the custom Tkinter window
    window = tk.Tk()
    window.title('DataFrame Viewer')
    
    # Create the Treeview widget with custom styling
    style = ttk.Style(window)
    style.theme_use("default")
    style.configure("Treeview",
                    background="#2a2d2e",
                    foreground="white",
                    rowheight=25,
                    fieldbackground="#343638",
                    bordercolor="#343638",
                    borderwidth=0)
    style.map('Treeview', background=[('selected', '#22559b')])
    style.configure("Treeview.Heading",
                    background="#565b5e",
                    foreground="white",
                    relief="flat")
    style.map("Treeview.Heading", background=[('active', '#3484F0')])
    
    # Update the function to display the table below the button
    def display_table():
        # Clear any previous table if it exists
        if 'treeview' in window.children:
            window.children['treeview'].destroy()
        
        # Create the Treeview widget
        treeview = ttk.Treeview(window, style="Treeview")
        treeview['columns'] = list(df.columns)
        
        # Format the Treeview columns
        for column in df.columns:
            treeview.heading(column, text=column, anchor=tk.CENTER)
            treeview.column(column, width=100)
        
        # Populate the Treeview with DataFrame data
        for i, row in df.iterrows():
            treeview.insert('', 'end', text=i, values=list(row))
        
        # Display the Treeview using the grid layout
        treeview.grid(row=1, column=0, padx=10, pady=10)
    
    # Create the button
    button = tk.Button(window, text='Show DataFrame', command=display_table)
    button.grid(row=0, column=0, padx=10, pady=10)
    
    # Start the Tkinter event loop
    window.mainloop()

# Call the function to display the DataFrame
show_dataframe()
```

```python
import customtkinter as ctk
from tkinter import ttk
import pandas as pd

def show_dataframe():
    # Create a sample DataFrame
    data = {'Name': ['John', 'Emma', 'Peter'],
            'Age': [25, 30, 35],
            'City': ['New York', 'London', 'Paris']}
    df = pd.DataFrame(data)
    
    # Create the custom Tkinter window
    window = ctk.CTk()
    window.title('Custom DataFrame Viewer')
    
    # Create the CTkTextbox widget
    textbox = ctk.CTkTextbox(window)
    textbox.pack(padx=10, pady=10)
    
    # Create the Treeview widget with custom styling
    style = ttk.Style(window)
    style.theme_use("default")
    style.configure("Treeview",
                    background="#2a2d2e",
                    foreground="white",
                    rowheight=25,
                    fieldbackground="#343638",
                    bordercolor="#343638",
                    borderwidth=0)
    style.map('Treeview', background=[('selected', '#22559b')])
    style.configure("Treeview.Heading",
                    background="#565b5e",
                    foreground="white",
                    relief="flat")
    style.map("Treeview.Heading", background=[('active', '#3484F0')])
    
    # Create the Treeview widget
    treeview = ttk.Treeview(window, style="Treeview")
    treeview['columns'] = list(df.columns)
    
    # Format the Treeview columns
    for column in df.columns:
        treeview.heading(column, text=column, anchor=ctk.CENTER)
        treeview.column(column, width=100)
    
    # Populate the Treeview with DataFrame data
    for i, row in df.iterrows():
        treeview.insert('', 'end', text=i, values=list(row))
    
    # Get the table text
    table_text = ""
    for column in df.columns:
        table_text += f"{column}\t"
    table_text += "\n"
    for i, row in df.iterrows():
        for value in row:
            table_text += f"{value}\t"
        table_text += "\n"
    
    # Set the table text in the CTkTextbox
    textbox.insert(text=table_text[0])
    
    # Start the custom Tkinter event loop
    window.run()

# Call the function to display the DataFrame
show_dataframe()
```

```python
import tkinter as tk
from tkinter import filedialog
import pyvistaqt as pvqt

def open_stl_file():
    # Open file dialog to select an STL file
    file_path = filedialog.askopenfilename(filetypes=[("STL Files", "*.stl")])

    # Load the STL file using PyVista
    mesh = pvqt.read(file_path)

    # Clear any existing mesh in the viewer
    viewer.clear()

    # Add the new mesh to the viewer
    viewer.add_mesh(mesh, color='white')

# Create the Tkinter window
window = tk.Tk()
window.title("STL Viewer")

# Create a button to open the STL file
open_button = tk.Button(window, text="Open STL File", command=open_stl_file)
open_button.pack()

# Create a PyVistaQt viewer
viewer = pvqt.BackgroundPlotter(window_size=(800, 600))
viewer.window.app.widget.pack(fill=tk.BOTH, expand=tk.YES)

window.mainloop()
```
